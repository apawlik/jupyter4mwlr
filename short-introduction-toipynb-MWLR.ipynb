## Interactive reporting

```python
from ipywidgets import interact
%matplotlib inline
```

```python
def func_plot(k=1):
    x = np.linspace(-2, 2, 200)
    plt.plot(x, np.sin(2*np.pi*k*x))
```

```python
interact(func_plot, k=(1,10))
```
